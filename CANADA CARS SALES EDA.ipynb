### Handling outliers

```python
# The lines of code print the 10th and 90th percentiles of the variable 'Sumofsales', respectively. 
# These values will be used for quantile-based flooring and capping.

print(car_sale['Sumofsales'].quantile(0.10))
print(car_sale['Sumofsales'].quantile(0.90))
```

```python
car_sale['Sumofsales'] = np.where(car_sale['Sumofsales'] <332.5, 332.5,car_sale['Sumofsales'])
car_sale['Sumofsales'] = np.where(car_sale['Sumofsales'] >9495.0, 9495.0,car_sale['Sumofsales'])
print(car_sale['Sumofsales'].skew())
```

```python
# updated boxplot after working on outliers
sns.boxplot(car_sale.Sumofsales)
```
